## 📘 Zustandsgrößen

- ***Quantifizierbare makroskopische Eigenschaften eines Systems***
- Durch die Zustandsgr¨ oßen wird der Zustand eines Systems eindeutig und un
abh¨ angig von der Vorgeschichte beschrieben.

### 🧩 Extensive Größen

Abhängig von der Systemgröße, additiv bei Teilsystemen

| Größe          | Symbol | Einheit | Typ      |
| -------------- | ------ | ------- | -------- |
| Volumen        | V      | m³      | extensiv |
| Masse          | m      | kg      | extensiv |
| Stoffmenge     | n      | mol     | extensiv |
| Innere Energie | U      | J       | extensiv |
| Enthalpie      | H      | J       | extensiv |
| Entropie       | S      | J/K     | extensiv |


### 🌡️ Intensive Größen

Unabhängig von der Systemgröße, bleiben konstant bei Teilung

| Größe         | Symbol | Einheit | Typ      |
| ------------- | ------ | ------- | -------- |
| Temperatur    | T      | K       | intensiv |
| Druck         | p      | Pa      | intensiv |
| Dichte        | ρ      | kg/m³   | intensiv |
| Konzentration | c      | mol/m³  | intensiv |


### ⚖️ Spezifische und molare Größen

Extensive Größen, normiert auf Masse oder Stoffmenge → intensiv

| Größe                      | Symbol   | Einheit  | Typ      |
| -------------------------- | -------- | -------- | -------- |
| Spezifisches Volumen       | v = V/m  | m³/kg    | intensiv |
| Spezifische innere Energie | u = U/m  | J/kg     | intensiv |
| Spezifische Enthalpie      | h = H/m  | J/kg     | intensiv |
| Spezifische Entropie       | s = S/m  | J/(kg·K) | intensiv |
| Molarvolumen               | Vm = V/n | m³/mol   | intensiv |



### 💡 Hinweis

* Extensive Größen → mit Systemgröße skalierend
* Intensive Größen → unabhängig von Systemgröße
* Spezifisch/molar → durch Normierung von extensiv → intensiv

## 📘 Thermodynamische Systeme

### 🧩 Haupttypen

| Systemtyp       | Stoffaustausch  | Energieaustausch | Beispiel            |
| --------------- | --------------- | ---------------- | ------------------- |
| **Offen**       | ✔️ möglich      | ✔️ möglich       | Turbine, Kompressor |
| **Geschlossen** | ❌ nicht möglich | ✔️ möglich       | Kolben mit Gas      |
| **Isoliert**    | ❌ nicht möglich | ❌ nicht möglich  | Thermosflasche      |



### 🌡️ Weitere Klassifikationen

| Eigenschaft       | Erklärung                                          | Beispiel                 |
| ----------------- | -------------------------------------------------- | ------------------------ |
| **Adiabat**       | kein Wärmeaustausch (Q = 0), nur Arbeit            | ideal isolierter Kolben  |
| **Nicht adiabat** | Wärmeaustausch möglich                             | Kochtopf ohne Isolierung |
| **Homogen**       | gleiche Zusammensetzung & Eigenschaften überall    | Glas voll Wasser         |
| **Heterogen**     | mehrere Phasen oder ungleichmäßige Zusammensetzung | Wasser + Luftblase       |
| **Reiner Stoff**  | homogen, eine Komponente                           | destilliertes Wasser     |



### ⚖️ Systemgrenzen

| Grenze          | Durchlässigkeit         | Bedeutung              |
| --------------- | ----------------------- | ---------------------- |
| **Diatherm**    | ✔️ Wärme                | Wärmeaustausch möglich |
| **Adiabat**     | ❌ Wärme                 | keine Wärmeübertragung |
| **Permeabel**   | ✔️ Stoff                | Stoffaustausch möglich |
| **Impermeabel** | ❌ Stoff                 | kein Stoffaustausch    |
| **Starr**       | kein Volumenänderung    | fester Behälter        |
| **Beweglich**   | Volumenänderung möglich | Kolben                 |



### 💡 Hinweis

* Offen = Stoff + Energie
* Geschlossen = nur Energie
* Isoliert = weder Stoff noch Energie
* Adiabat = kein Q, nur W

Mit dieser Tabelle hast du eine klare Übersicht:

* **Drei Grundsysteme** (offen, geschlossen, isoliert)
* **Zusatzklassifikationen** (adiabat, homogen …)
* **Systemgrenzen** regeln Austauschmöglichkeiten

## 📘 Thermodynamisches Gleichgewicht

### 🧩 Hauptarten des Gleichgewichts

| Gleichgewicht  | Bedeutung                                   | Symbolisch | Beispiel                             |
| -------------- | ------------------------------------------- | ---------- | ------------------------------------ |
| **Mechanisch** | keine unausgeglichenen Kräfte               | Δp = 0     | gleichmäßiger Druck in Flüssigkeit   |
| **Thermisch**  | keine Temperaturunterschiede                | ΔT = 0     | Wasser im Becher überall gleich warm |
| **Chemisch**   | keine spontanen Reaktionen, keine Diffusion | Δμ = 0     | stabile Gas-Mischung ohne Reaktion   |



### 🌡️ Thermisches Gleichgewicht und 0. Hauptsatz

* Zwei Systeme im thermischen Kontakt gleichen ihre Temperatur an → ΔT = 0.
* **0. Hauptsatz**: Wenn A mit B und B mit C im thermischen Gleichgewicht stehen, dann auch A mit C.



### ⚖️ Thermodynamisches Gleichgewicht

* Liegt vor, wenn **alle drei Teilbedingungen** gleichzeitig erfüllt sind:

  * Mechanisch
  * Thermisch
  * Chemisch
* Nur dann sind **alle Zustandsgrößen konstant**, solange das System isoliert ist.



### 💡 Hinweis

* Thermisch = nur Temperaturausgleich
* Thermodynamisch = vollständiger Gleichgewichtszustand (mechanisch + thermisch + chemisch)
* Einfache Systeme: durch **2 unabhängige Größen** vollständig beschreibbar (z. B. p und ρ).


### ✅ Zusammenfassung

* **3 Arten von Gleichgewicht:** mechanisch, thermisch, chemisch
* **Thermodynamisches Gleichgewicht = alle drei erfüllt**
* **0. Hauptsatz** = Temperaturausgleich als Bedingung für thermisches Gleichgewicht

## 📘 Thermisches Gleichgewicht
### 🧩 Definition
* Thermisches Gleichgewicht liegt vor, wenn zwei Systeme oder ein System und seine Umgebung keine netto Wärme austauschen. Dies geschieht, wenn die Temperaturen gleich sind und der Energieaustausch im Gleichgewicht ist.

## 📘 Hauptsätze der Thermodynamik

### 🧩 Nullter Hauptsatz

| Hauptsatz | Aussage                                                                        | Formel               | Bemerkung                              |
| --------- | ------------------------------------------------------------------------------ | -------------------- | -------------------------------------- |
| **0. HS** | Wenn A mit B und B mit C im thermischen Gleichgewicht steht, dann auch A mit C | A ⇌ B, B ⇌ C ⇒ A ⇌ C | Grundlage der **Temperaturdefinition** |

### 1️⃣ Erster Hauptsatz – Energieerhaltung

| Systemtyp                   | Formulierung                                             | Formel                        | Bemerkung                                         |
| --------------------------- | -------------------------------------------------------- | ----------------------------- | ------------------------------------------------- |
| **Allgemein (geschlossen)** | Änderung der inneren Energie = zugeführte Wärme + Arbeit | ΔU = Q + W                    | Energie kann weder erzeugt noch vernichtet werden |
| **Differentiell**           | spezifisch                                               | du = dq + dw                  |                                                   |
| **Isochor (V = const.)**    | keine Volumenarbeit                                      | ΔU = Q                        | Q = ΔU                                            |
| **Isobar (p = const.)**     | Enthalpie als Zustandsgröße                              | ΔH = Q                        | Q = ΔH                                            |
| **Adiabat**                 | kein Wärmeaustausch (Q = 0)                              | ΔU = W                        | nur Arbeit ändert innere Energie                  |
| **Isoliert**                | keine Wechselwirkung mit Umgebung                        | ΔU = 0                        | Energie bleibt konstant                           |
| **Offenes System**          | Erweiterung durch Massen- und Strömungsenergie           | ΔE = Q + W + (Ez,in − Ez,out) | Energie kann mit Stoff transportiert werden       |


### 2️⃣ Zweiter Hauptsatz – Entropie & Richtung

| Systemtyp                | Formulierung                                             | Formel      | Bemerkung                       |
| ------------------------ | -------------------------------------------------------- | ----------- | ------------------------------- |
| **Allgemein**            | Wärme kann nicht vollständig in Arbeit verwandelt werden | ∮ δQ/T ≤ 0  | Prozess hat bevorzugte Richtung |
| **Reversibel**           | Entropie konstant                                        | ΔS = 0      |                                 |
| **Irreversibel**         | Entropie nimmt zu                                        | ΔS > 0      |                                 |
| **Adiabat (reversibel)** | Isentropie                                               | ΔS = 0      | Beispiel: ideale Turbine        |
| **Nicht adiabat**        | Wärmeaustausch vorhanden                                 | ΔS ≥ ∫ δQ/T |                                 |
| **Isoliert**             | keine Energie- und Stoffströme                           | ΔS ≥ 0      | irreversibel: ΔS > 0            |

### 3️⃣ Dritter Hauptsatz – Absoluter Nullpunkt

| Hauptsatz          | Aussage                                                                           | Formel             | Bemerkung                      |
| ------------------ | --------------------------------------------------------------------------------- | ------------------ | ------------------------------ |
| **3. HS**          | Der absolute Nullpunkt T = 0 K kann nicht durch endliche Prozesse erreicht werden | lim T→0 S = const. | „Prinzip der Unerreichbarkeit“ |
| **Nernst-Theorem** | ΔS → 0 bei T → 0                                                                  | ΔS → 0             | gilt für reversible Prozesse   |


### 🌡️ Ergänzende Übersicht – Systeme

| Systemtyp         | Stoffaustausch | Wärmeaustausch | Arbeit | Formel                        |
| ----------------- | -------------- | -------------- | ------ | ----------------------------- |
| **Offen**         | ✔️             | ✔️             | ✔️     | ΔE = Q + W + ṁ(h + ½v² + gz) |
| **Geschlossen**   | ❌              | ✔️             | ✔️     | ΔU = Q + W                    |
| **Isoliert**      | ❌              | ❌              | ❌      | ΔU = 0, ΔS ≥ 0                |
| **Adiabat**       | abhängig       | ❌              | ✔️     | ΔU = W                        |
| **Nicht adiabat** | abhängig       | ✔️             | ✔️     | ΔU = Q + W                    |


### 💡 Hinweis

* **Adiabat = kein Q, nur W**
* **Isoliert = nix rein, nix raus (Energie und Stoff konstant)**
* **Offen = Stoff + Energieaustausch**
* **Geschlossen = kein Stoff, aber Energie möglich**


### ✅ Zusammenfassung

* **0. HS** → Temperaturdefinition
* **1. HS** → Energieerhaltung, Varianten: isochor, isobar, offen, isoliert, adiabat
* **2. HS** → Entropie & Richtung: reversibel/irreversibel, adiabate isentrop, isoliert ΔS ≥ 0
* **3. HS** → T = 0 K unerreichbar, Entropie → konst.

## 📘 Ideales Gasgesetz

### 🧩 Formen des Gasgesetzes

| Form                        | Formel              | Größen                |
| --------------------------- | ------------------- | --------------------- |
| **Spezifisch**              | $p , v = R , T$     | $v = \tfrac{1}{\rho}$ |
| **Massenbezogen**           | $p , V = m , R , T$ | $m =$ Masse           |
| **Molbezogen (universell)** | $p , V = n , R , T$ | $n =$ Stoffmenge      |


### 🌡️ Herleitung

* Boyle–Mariotte:
  $$
  p , v \big|_{T} = \text{konst.}
  $$
* Gay-Lussac:
  $$
  v \big|_{p} \sim T
  $$
* Kombination:
  $$
  p , v = R , T
  $$


### ⚖️ Gaskonstanten

* Universelle Gaskonstante:
  $$
  R = 8{,}314 \ \tfrac{\text{J}}{\text{mol} \cdot \text{K}}
  $$
* Spezifische Gaskonstante:
  $$
  R_s = \tfrac{R}{M}
  $$

  Beispiel Luft:
  $$
  R_s = 286{,}7 \ \tfrac{\text{J}}{\text{kg} \cdot \text{K}}, \quad M \approx 28{,}96 \ \tfrac{\text{g}}{\text{mol}}
  $$



### 📏 Verknüpfungen

* Formen: spezifisch, massenbezogen, molbezogen
* Konstanten:
  $$
  R = 8{,}314 \ \tfrac{\text{J}}{\text{mol} \cdot \text{K}}, \quad R_s = \tfrac{R}{M}
  $$
* Beispiel Luft:
  $$
  R_s = 286{,}7 \ \tfrac{\text{J}}{\text{kg} \cdot \text{K}}
  $$
* Wichtige Beziehungen:
  $$
  p = \rho \cdot R \cdot T,  \quad c_p - c_v = R, \quad \kappa = \tfrac{c_p}{c_v}
  $$



### 💡 Hinweis

* Drei gleichwertige Schreibweisen:
  $$
  p \cdot v = R \cdot T, \quad p \cdot V = m \cdot R \cdot T, \quad p \cdot V = n \cdot R \cdot T
  $$
* Universell = unabhängig vom Stoff
* Spezifisch = stoffabhängig (über Molmasse)
* Näherung gilt gut für ideale Gase bei normalen Drücken und Temperaturen



### ✅ Zusammenfassung

* **Formen:** spezifisch, massenbezogen, molbezogen
* **Konstanten:** $R = 8{,}314 \ \tfrac{\text{J}}{\text{mol} \cdot \text{K}}, \quad R_s = \tfrac{R}{M}$
* **Beispiel Luft:** $R_s = 286{,}7 \ \tfrac{\text{J}}{\text{kg} \cdot \text{K}}$
* **Wichtige Verknüpfungen:** $p = \rho R T, \quad c_p - c_v = R, \quad \kappa = \tfrac{c_p}{c_v}$

## 📘 Thermodynamische Prozesse 

### 🧩 Hauptprozesse

| Prozess      | Bedingung      | Hauptformel                                | Bemerkung               |
| ------------ | -------------- | ------------------------------------------ | ----------------------- |
| **Isotherm** | ( T = konst. ) | ( p V = \text{konst.} )                    | Wärme = Arbeit          |
|              |                | ( W = n R T \ln \tfrac{V_2}{V_1} )         |                         |
| **Isochor**  | ( V = konst. ) | ( W = 0 )                                  | Wärme = ΔU              |
|              |                | ( Q = \Delta U )                           |                         |
| **Isobar**   | ( p = konst. ) | ( W = p (V_2 - V_1) )                      | Wärme = ΔH              |
|              |                | ( Q = \Delta H )                           |                         |
| **Isentrop** | ( S = konst. ) | ( p V^\kappa = \text{konst.} )             | reversibel + adiabat    |
|              |                | ( T V^{\kappa-1} = \text{konst.} )         |                         |
|              |                | ( W = \tfrac{p_2 V_2 - p_1 V_1}{1-\kappa}) |                         |
| **Adiabat**  | ( Q = 0 )      | ( \Delta U = W )                           | Wärmestrom = 0          |


### 🌡️ Allgemeine Formeln

* Arbeit allgemein: ( W = \int p , dV )
* Wärme allgemein: ( Q = \int T , dS )
* Erster Hauptsatz: ( \Delta U = Q + W )


### 💡 Hinweis

* Isotherm: Temperatur konstant, ( Q = W )
* Isochor: Volumen konstant, ( Q = \Delta U )
* Isobar: Druck konstant, ( Q = \Delta H )
* Isentrop: Entropie konstant, ( p V^\kappa = konst. )
* Adiabat: kein Wärmeaustausch, ( \Delta U = W )


### ✅ Zusammenfassung

* Prozesse = Zustandsänderungen, eingeteilt nach Konstanz einer Größe
* Allgemeine Form: ( p V^n = konst. )
* Arbeit und Wärme über Integrale ( W = ∫p dV, Q = ∫T dS )
* Isentrop = Spezialfall von adiabatem, wenn reversibel

## 📘 p/v und T-S Diagramm

```python
# Volumen-Bereich
V = np.linspace(0.5, 3.0, 100)
n = 1.0   # mol
R = 8.314
T_iso = 1
p_iso = n*R*T_iso/V

# Isotherm (T = konst.)
p_isotherm = p_iso

# Isobar (p = konst.)
p_isobar = np.ones_like(V)*1.0

# Isochor (V = konst.)
V_isochor = np.ones(200)*1.0
p_isochor = np.linspace(0.5,15,200)

# Isentrop (p*V^kappa=konst.)
kappa = 1.4
p_isentrop = p_iso[0]*V[0]**kappa / V**kappa

# Plot p-V Diagramm
plt.figure(figsize=(8,6))
plt.plot(V, p_isotherm, label="Isotherm (T=konst.)")
plt.plot(V, p_isentrop, label="Isentrop (S=konst.)")
plt.plot(V, p_isobar, label="Isobar (p=konst.)")
plt.plot(V_isochor, p_isochor, label="Isochor (V=konst.)")

plt.xlabel("Volumen V [m³]")
plt.ylabel("Druck p [bar, normiert]")
plt.title("p-V-Diagramm: Thermodynamische Prozesse")
plt.legend()
plt.grid(True)
plt.show()
```

```python
s = np.linspace(0, 3, 200)
T0 = 300

# Isotherm: T = konst
T_isotherm = np.ones_like(s)*T0

# Isobar: T steigt stärker mit Entropie, leichte Krümmung
T_isobar = T0 + 120*(np.sqrt(s+1)-1)

# Isochor: T steigt noch stärker, deutlichere Krümmung
T_isochor = T0 + 300*(np.sqrt(s+1)-1)

# Isentrop: s = konst
s_isentrop = np.ones(200)*1.5
T_isentrop = np.linspace(250, 600, 200)

# Plot T-s Diagramm
plt.figure(figsize=(8,6))
plt.plot(s, T_isotherm, label="Isotherm (T=konst.)", linewidth=2)
plt.plot(s, T_isobar, label="Isobar (p=konst.)", linewidth=2)
plt.plot(s, T_isochor, label="Isochor (V=konst.)", linewidth=2)
plt.plot(s_isentrop, T_isentrop, label="Isentrop (S=konst.)", linewidth=2)

plt.xlabel("Entropie s [kJ/(kg K)] (schematisch)")
plt.ylabel("Temperatur T [K]")
plt.title("T-s-Diagramm: Thermodynamische Prozesse")
plt.legend()
plt.grid(True)
plt.show()
```

## 📘 Erster Hauptsatz der Thermodynamik

### 🧩 Allgemeine Form

Der **1. Hauptsatz** beschreibt die **Energieerhaltung**:

$$
\Delta U = Q + W
$$

mit

* $U$ = innere Energie
* $Q$ = zugeführte Wärme
* $W$ = zugeführte Arbeit (inkl. Volumenänderungsarbeit, technische Arbeit etc.)


### 🌡️ Geschlossene Systeme

Für ein ruhendes, geschlossenes System:

$$
dU = \delta Q + \delta W
$$

* **Differentiell, spezifisch:**
  $$
  du = \delta q + \delta w
  $$

* **Bei quasistatischer Zustandsänderung:**
  $$
  dU = \delta Q - p \cdot dV
  $$

* **Sonderfälle:**

  * Isochor ($dV = 0$):
    $$dU = \delta Q$$
  * Isobar:
    $$dH = \delta Q$$


### ⚖️ Offene Systeme

Hier muss zusätzlich der **Energieaustausch durch Massenströme** berücksichtigt werden.

Allgemein:

$$
dE = \delta Q + \delta W + d^{(m)}E
$$

mit $d^{(m)}E$ = Energie durch Massenstrom (Transport von innerer, kinetischer und potentieller Energie).

* Aufgeteilt:

$$
dU = \delta Q + \delta W_t + p_z v_z , dm_z - p_a v_a , dm_a + d^{(m)}U
$$

$z =$ Zustrom, $a =$ Abstrom.

* **Stationärer Fall (Energiegleichung):**

$$
0 = \dot Q + \dot W_t + \sum \dot m \left(h + \frac{c^2}{2} + gz \right)*{\text{ein}} - \sum \dot m \left(h + \frac{c^2}{2} + gz \right)*{\text{aus}}
$$

Dies ist die **Energiegleichung für stationäre offene Systeme** (z. B. Turbine, Verdichter, Wärmetauscher).


### 📏 Abwandlungen

* **Differentialform:**
  $$
  dU = \delta Q + \delta W
  $$

* **Spezifisch (pro Masse):**
  $$
  du = \delta q + \delta w
  $$

* **Enthalpie-Form (für offene Systeme, stationär):**
  $$
  \dot Q + \dot W_t = \sum \dot m , h_{\text{ein}} - \sum \dot m , h_{\text{aus}}
  $$



### ✅ Zusammenfassung

* Geschlossen: **Innere Energieänderung = Wärme + Arbeit**.
* Offenes System: **Zusätzlich Energie durch Massenströme (Enthalpie, kinetisch, potenziell)**.
* Stationär: **Energieerhaltung in Bilanzgleichung (oft mit Enthalpie ausgedrückt)**.

### 💡 Hinweis

Im geschlossenen System ändern nur **$Q$ und $W$** die Energie.
Im offenen System kommt zusätzlich der **konvektive Transport** mit der Masse dazu → deshalb wird oft mit **Enthalpie** gerechnet.

## 📘 Arbeit in der Thermodynamik

### 🧩 Übersicht

| Art der Arbeit             | Beschreibung                                                                                                                    | Formel mit Integrationsgrenzen                                                   |
| -------------------------- | ------------------------------------------------------------------------------------------------------------------------------- | -------------------------------------------------------------------------------- |
| **Volumenänderungsarbeit** | Arbeit durch Expansion oder Kompression eines Gases. Fläche im p-V-Diagramm.                                                    | $$ W_{12} = - \int_{V_1}^{V_2} p \cdot dV $$                                         |
| **Elektrische Arbeit**     | Arbeit durch Stromfluss über Spannung oder Widerstand. Immer positiv (Energiezufuhr).                                           | $$ W_{el,12} = \int_{t_1}^{t_2} U I , dt = \int_{t_1}^{t_2} R I^2 \cdot dt $$        |
| **Verschiebearbeit**       | Arbeit in offenen Systemen durch Massestrom an den Systemgrenzen (z. B. Düsen\cdot Kompressoren).                                   | $$ W_p = \int_{m_1}^{m_2} p v \cdot dm \quad \text{mit} \quad dW_p = \pm p v \cdot dm $$ |
| **Technische Arbeit**      | Arbeit in Strömungsmaschinen (Pumpen, Turbinen, Verdichtern). Kombination aus Enthalpie-, kinetischer und potenzieller Energie. | $$ W_t = (h_1 - h_2) + \frac{c_1^2 - c_2^2}{2} + g (z_1 - z_2) $$                |
| **Mechanische Arbeit**     | Grundformel: Energieübertragung durch eine Kraft längs eines Weges.                                                             | $$ W_{12} = \int_{s_1}^{s_2} F \cdot ds $$                                           |
| **Arbeit im Kreisprozess** | Gesamtarbeit eines Zyklus. Fläche, die der Prozess im p-V-Diagramm einschließt.                                                 | $$ W = \oint p \cdot dV $$                                                           |


### 🌡️ Erläuterungen

* **Volumenänderungsarbeit**
  $$ W_{12} = - \int_{V_1}^{V_2} p \cdot dV $$
  Vorzeichen: Expansion → Arbeit an Umgebung (negativ), Kompression → Arbeit am System (positiv).

* **Elektrische Arbeit**
  $$ W_{el,12} = \int_{t_1}^{t_2} U I \cdot dt = \int_{t_1}^{t_2} R I^2 \cdot dt $$
  Immer positiv, da Widerstände Energie nur aufnehmen.

* **Verschiebearbeit**
  $$ W_p = \int_{m_1}^{m_2} p v \cdot dm \quad \text{mit} \quad dW_p = \pm p v \cdot dm $$

* **Technische Arbeit**
  $$ W_t = (h_1 - h_2) + \frac{c_1^2 - c_2^2}{2} + g (z_1 - z_2) $$

* **Mechanische Arbeit**
  $$ W_{12} = \int_{s_1}^{s_2} F \cdot ds $$

* **Arbeit im Kreisprozess**
  $$ W = \oint p \cdot dV $$


### 📏 Verknüpfungen

Alle Arbeitsformen basieren auf dem Grundprinzip **Kraft × Weg** bzw. Energieübertragung über Systemgrenzen.
Sie stehen in direkter Verbindung mit dem **1. Hauptsatz**:

$$
\Delta U = Q + W
$$

### 💡 Hinweis

Vorzeichen beachten:

* **Arbeit am System** → positiv
* **Arbeit vom System an Umgebung** → negativ


### 📖 Zusammenfassung

Die wichtigsten Arbeitsarten in der Thermodynamik sind:

* Volumenänderungsarbeit (Expansion/Kompression)
* Elektrische Arbeit (durch Stromfluss)
* Verschiebearbeit (bei offenen Systemen)
* Technische Arbeit (z. B. in Turbinen)

Alle haben gemeinsam, dass sie **Energieübertragung durch gerichtete Prozesse** beschreiben, im Gegensatz zur Wärme, die durch **Temperaturdifferenzen** übertragen wird.

* Volumenänderungsarbeit:
  $$ W_{12} = - \int_{V_1}^{V_2} p \cdot dV $$

* Elektrische Arbeit:
  $$ W_{el12} = \int_{t_1}^{t_2} U I \cdot dt $$

* Verschiebearbeit:
  $$ W_p = \int_{m_1}^{m_2} p v \cdot dm $$

* Mechanische Arbeit:
  $$ W_{12} = \int_{s_1}^{s_2} F \cdot ds $$

* Technische Arbeit:
  $$ W_t = W_m + W_v + W_el $$

## 📘 Arbeit bei festen Systemgrenzen

### 🧩 Grundidee

Arbeit in der Thermodynamik entsteht, wenn sich **Systemgrenzen verschieben** und dabei Kräfte Wege zurücklegen (z. B. Kolben bewegt sich).

### 🌡️ Begründung

* Mechanische Arbeit wird allgemein über das Integral

$$
W = \int p , dV
$$

beschrieben.

* Wenn das **Volumen konstant bleibt** (keine Bewegung der Systemgrenzen, $dV = 0$), dann verschwindet auch das Integral → es wird **keine Arbeit verrichtet**.
* Auch beim Überströmen von Gasen durch ein Ventil gilt: Da sich das Volumen der einzelnen Behälter nicht ändert, wird keine technische Arbeit verrichtet.

### ⚖️ Bezug zum Ersten Hauptsatz

Der Erste Hauptsatz lautet:

$$
\Delta U = Q + W
$$

Wenn $W = 0$, können Änderungen der inneren Energie nur durch **Wärmeübertragung $Q$** entstehen.

### 📏 Verknüpfung

* Arbeit ≠ Zustandsgröße, sondern Prozessgröße.
* Nur wenn eine Kraft über eine Verschiebung wirkt, ist Arbeit definiert.
* Feste Systemgrenzen ⇒ kein Weg ⇒ keine Arbeit.

### ✅ Zusammenfassung

Wenn sich die **Systemgrenzen nicht verschieben**, gibt es keinen Weg für die Kraft, also keine mechanische Arbeit. Damit ist die Arbeit $W = 0$, selbst wenn sich Druck oder Temperatur im Inneren ändern.

### 💡 Hinweis

Merke:
Arbeit in der Thermodynamik ist immer „Kraft mal Weg“ → **kein Weg = keine Arbeit**.

## 📘 Erster Hauptsatz – offene Systeme

### 🧩 Geschlossene Systeme

* **Kein Stoffaustausch**, nur Energieaustausch (Wärme, Arbeit).
* Gesamtenergieänderung = Wärme + Arbeit:

( dU = dQ + dW )

Bei quasistatischer Zustandsänderung:

( dU = dQ - p dV )

### 🌡️ Offene Systeme

* **Stoff- und Energieaustausch** über Systemgrenzen.
* Zusätzlich zu Wärme und Arbeit wird Energie mit der Materie transportiert:

  * innere Energie (u)
  * kinetische Energie (v²/2)
  * potentielle Energie (gz)
  * Verschiebearbeit (p v)

Erweiterter 1. Hauptsatz:

( dE = dQ + dW_t + (h + v²/2 + gz) dm )

oder integriert:

( U₂ - U₁ = Q₁₂ + W_t,₁₂ + (h_z - h_a) )

mit h = Enthalpie (u + p v).



### ⚖️ Kerndifferenz

| Aspekt             | Geschlossenes System | Offenes System                                                   |
| ------------------ | -------------------- | ---------------------------------------------------------------- |
| Stoffaustausch     | Nein                 | Ja                                                               |
| Energieübertragung | Wärme & Arbeit       | Wärme, technische Arbeit, Verschiebearbeit, mitströmende Energie |
| Bilanzgleichung    | ( ΔU = Q + W )       | ( ΔU = Q + W_t + Σ m(h+v²/2+gz) )                                |
| Anwendung          | Kolben, Zylinder     | Turbine, Pumpe, Düse, Wärmetauscher                              |



### 📖 Zusammenfassung

Der **Unterschied** liegt darin, dass beim **geschlossenen System** nur Wärme und Arbeit bilanziert werden, während beim **offenen System** zusätzlich die mitströmende Materie Energie transportiert (Enthalpie, kinetische und potentielle Energie).

## 📘 Enthalpie

### 🧩 Definition

| Größe                 | Formel            | Bemerkung          |
| --------------------- | ----------------- | ------------------ |
| Enthalpie             | H = U + pV        | extensive          |
| Spezifische Enthalpie | h = u + p/ρ       | intensive          |
| Ideales Gas           | h(T) = u(T) + R·T | nur von T abhängig |



### 🌡️ Zusammenhang mit Prozessen

* Isochor (V = konst.)
  Q = ΔU
  → gesamte Wärme geht in innere Energie

* Isobar (p = konst.)
  Q = ΔH
  → gesamte Wärme geht in Enthalpie

* Allgemein (quasistatisch):
  dH = dQ + V dp



### ⚖️ Wichtige Anwendungen

* **Drosselvorgang (Joule-Thomson):**
  h1 = h2 (bei vernachlässigbarer kinetischer/potentieller Energie)
  – ideales Gas → T bleibt konstant
  – reales Gas → T kann steigen oder sinken

* **Energie- und Stoffströme in offenen Systemen:**
  Enthalpie beschreibt die mitströmende Energie pro Masseneinheit.



### 📌 Hinweise für händische Berechnungen

1. Prozessart prüfen:

   * Isochor → Δu = q
   * Isobar → Δh = q

2. Bei idealen Gasen gilt:

   * h = cp·T
   * Δh = cp·ΔT

3. Immer unterscheiden zwischen **spezifischen Größen** (h, u) und **Gesamtgrößen** (H, U).

4. Konstanten (cp, cv, κ) können temperaturabhängig sein → im Skript meist konstant gesetzt.

5. Für reale Gase: h kann druckabhängig sein → Tabellen/Werte nötig.

6. Bei dp ≠ 0 nicht vergessen:
   dH = dQ + V dp



### ✅ Merktabelle

| Prozess                    | Beziehung           | Bemerkung                                     |
| -------------------------- | ------------------- | --------------------------------------------- |
| Isochor (V = konst.)       | q = Δu              | Wärme = Änderung innere Energie               |
| Isobar (p = konst.)        | q = Δh              | Wärme = Änderung Enthalpie                    |
| Drosselung (p↓)            | h1 = h2             | Ideales Gas: T = konst.                       |
| Offenes System (stationär) | ṁ·h Ein = ṁ·h Aus | Enthalpie beschreibt Energie mit der Strömung |



### 💡 Hinweis

* Enthalpie ist eine **Hilfsgröße**, die Rechnungen bei offenen Systemen und isobaren Prozessen stark vereinfacht.
* Bei idealen Gasen genügt oft h = cp·T.
* Bei realen Gasen Tabellenwerte nutzen und Druckabhängigkeit beachten.



### ✅ Zusammenfassung

* **Definition:** H = U + pV, h = u + p/ρ
* **Ideales Gas:** h = cp·T, nur temperaturabhängig
* **Isobar:** Q = ΔH
* **Isochor:** Q = ΔU
* **Besonders achten:** Prozessart, spezifische vs. gesamte Größen, reale vs. ideale Gase, Druckabhängigkeit

## 📘 Spezifische Wärmekapazitäten

### 🧩 Definitionen

Die spezifischen Wärmekapazitäten beschreiben, wie viel Wärme pro Masseneinheit benötigt wird, um die Temperatur um 1 K zu erhöhen:

* Spezifische Wärmekapazität bei konstantem Volumen:
  $$
  c_v = \left(\frac{\partial u}{\partial T}\right)_v
  $$

* Spezifische Wärmekapazität bei konstantem Druck:
  $$
  c_p = \left(\frac{\partial h}{\partial T}\right)_p
  $$

* Adiabatenexponent (Isentropenexponent):
  $$
  \kappa = \frac{c_p}{c_v}
  $$

### 🌡️ Herleitung

1. Ausgangspunkt: Enthalpie-Definition
   $$
   h = u + pv
   $$

2. Für ideale Gase mit $pv = RT$:
   $$
   h = u + RT
   $$

3. Ableitung nach der Temperatur:
   $$
   \frac{dh}{dT} = \frac{du}{dT} + R
   $$

4. Mit den Definitionen von $c_p$ und $c_v$:
   $$
   c_p = c_v + R
   $$

### 📏 Wichtige Beziehungen

* Mayer-Beziehung:
  $$
  c_p - c_v = R
  $$


### 🔬 Physikalische Interpretation

* **$c_v$**: Energiezufuhr erhöht nur die innere Energie (Translationsenergie der Teilchen)
* **$c_p$**: Zusätzliche Energie nötig für Volumenarbeit gegen konstanten Druck
* **$\kappa$**: Maß für die "Steifigkeit" des Gases bei adiabaten Prozessen

<hr style="border: none; border-top: 1px solid #2ecc71;">

## 📘 Zusammenhang spezifische und molare Wärmekapazität

### 1. Start: Definition der spezifischen und molaren Wärmekapazität

* **Spezifische Wärmekapazität bei konstantem Volumen**:

$$
c_v = \left( \frac{\partial u}{\partial T} \right)_v
$$

mit $u = U/m$ (innere Energie pro kg).

* **Molare Wärmekapazität bei konstantem Volumen**:

$$
C_v = \left( \frac{\partial \bar{u}}{\partial T} \right)_v
$$

mit $\bar{u} = U/n$ (innere Energie pro Mol).



### 2. Gesamtenergieänderung über die Masse oder Molzahl

* Über die **Masse**:

$$
dU = m \, du = m \, c_v \, dT
$$

* Über die **Stoffmenge**:

$$
dU = n \, d\bar{u} = n \, C_v \, dT
$$



### 3. Zusammenhang zwischen spezifisch und molar

Da $n = \tfrac{m}{M}$ (M = molare Masse), gilt:

$$
c_v = \frac{C_v}{M}
$$

also sind beide Definitionen konsistent.



### 4. Integration für konstantes $c_v, C_v$

* Mit **Masse**:

$$
\Delta U = m \, c_v \, (T_2 - T_1)
$$

* Mit **Molzahl**:

$$
\Delta U = n \, C_v \, (T_2 - T_1)
$$



### 5. Mathematisches „Wegkürzen“

Wir nehmen die molare Form:

$$
\Delta U = n \, C_v \, \Delta T
$$

Nun zeigen wir, wie das aus der Masse-Form folgt:

$$
\Delta U = m \, c_v \, \Delta T
$$

Einsetzen von $c_v = \tfrac{C_v}{M}$:

$$
\Delta U = m \cdot \frac{C_v}{M} \cdot \Delta T
$$

Da $\tfrac{m}{M} = n$, kürzt sich:

$$
\Delta U = \underbrace{\frac{m}{M}}_{=n} \cdot C_v \cdot \Delta T
$$

$$
\Delta U = n \, C_v \, \Delta T
$$

## 📘 Isentrope Zustandsänderung

### 🧩 Formen des Gesetzes

Für ein ideales Gas mit konstanter Wärmekapazität und dem Adiabatenexponenten $\kappa = \tfrac{c_p}{c_v}$ gilt:

* Temperatur-Volumen-Beziehung:
  $$
  T v^{\kappa - 1} = \text{konst.}
  $$

* Druck-Volumen-Beziehung (Poisson-Gleichung):
  $$
  p v^\kappa = \text{konst.}
  $$

* Temperatur-Druck-Beziehung:
  $$
  \frac{T_2}{T_1} = \left(\frac{p_2}{p_1}\right)^{\tfrac{\kappa - 1}{\kappa}}
  $$

### 🌡️ Herleitung

1. Ausgangspunkt: erster Hauptsatz für ein geschlossenes System
   $$
   dU = \delta Q - \delta W
   $$

2. Für quasistatische, adiabate Prozesse:
   $$
   \delta Q = 0 \quad \Rightarrow \quad dU = -p \cdot dV
   $$

3. Mit $dU = c_v dT$ und idealem Gasgesetz $pV = RT$:
   $$
   c_v dT = - \frac{RT}{V} dV
   $$

4. Umformen:
   $$
   \frac{dT}{T} = -(\kappa - 1) \frac{dV}{V}
   $$

5. Integrieren:
   $$
   \ln T + (\kappa - 1)\ln V = \text{konst.}
   $$

   $$
   T V^{\kappa - 1} = \text{konst.}
   $$

6. Mit idealem Gasgesetz:
   $$
   p V^\kappa = \text{konst.}
   $$

### 🌡️ Ergebnisse:
$$
p v^\kappa = \text{konst.}
$$
$$
T v^{\kappa - 1} = \text{konst.}
$$
$$
T p^{\tfrac{1-\kappa}{\kappa}} = \text{konst.}
$$
### ⚖️ Annahmen

* Ideales Gas (keine Wechselwirkungen zwischen den Teilchen außer elastische Stöße)
* Quasistatischer Prozess (immer im Gleichgewichtszustand)
* Adiabater Prozess (kein Wärmeaustausch: $\delta Q = 0$)
* Konstante spezifische Wärmekapazitäten $c_p, c_v$ → damit $\kappa$ konstant
* Keine dissipativen Effekte (Reibung, Wärmeleitung, Viskosität vernachlässigt)

### 📏 Verknüpfungen

* Zusammenhang zwischen isentrop und adiabatisch:
  Jede quasistatische adiabate Zustandsänderung ist isentrop ($ds = 0$).
  Umgekehrt ist nicht jede isentrope Änderung adiabat (z.B. isotherme Entropieerhaltung mit Wärmeaustausch).

* Isentrope Prozesse sind wichtig für Turbinen, Verdichter, Düsen (Modellannahme).

### ✅ Zusammenfassung

Die isentrope Zustandsänderung folgt aus der Kombination von:

* Erstem Hauptsatz mit $\delta Q = 0$
* Idealer Gasgleichung
* Annahme konstanter Wärmekapazitäten



### 💡 Hinweis

Die Isentrope beschreibt einen **idealen Grenzprozess**. In der Realität gibt es immer Irreversibilitäten → reale Prozesse sind nur **näherungsweise isentrop** (z.B. Wirkungsgrade bei Turbinen und Verdichtern).

## 🎯 Aufgaben Kapitel 2 

### ✅ 2.4 Überstromprozess

#### 🧩 Ausgangspunkt

Die gezeigte Gleichung lautet:

$$
\frac{m_{A2}}{m_{A1}} = \frac{v_{A1}}{v_{A2}} = \left( \frac{p_{A2}}{p_{A1}} \right)^{\frac{1}{\kappa}}
$$

Sie beschreibt, **wie sich Masse, spezifisches Volumen und Druck** in einem **isentropen Prozess** zueinander verhalten.



#### 🌡️ Herleitung

1. **Isentrope Beziehung für p und v:**
   $$
   p , v^{\kappa} = \text{konst.}
   $$
   → Für zwei Zustände A1 und A2:
   $$
   \frac{p_{A2}}{p_{A1}} = \left( \frac{v_{A1}}{v_{A2}} \right)^{\kappa}
   $$

2. **Umgestellt nach den spezifischen Volumina:**
   $$
   \frac{v_{A1}}{v_{A2}} = \left( \frac{p_{A2}}{p_{A1}} \right)^{\frac{1}{\kappa}}
   $$

3. **Bezug zur Masse:**
   Das spezifische Volumen $v$ ist der Kehrwert der Dichte $\rho$:
   $$
   v = \frac{1}{\rho} \quad \Rightarrow \quad \rho = \frac{1}{v}
   $$
   Und für ein gegebenes Volumen $V$:
   $$
   m = \rho V = \frac{V}{v}
   $$
   Wenn das Volumen des Systems gleich bleibt, folgt:
   $$
   \frac{m_{A2}}{m_{A1}} = \frac{v_{A1}}{v_{A2}}
   $$

4. **Einsetzen ergibt:**
   $$
   \frac{m_{A2}}{m_{A1}} = \frac{v_{A1}}{v_{A2}} = \left( \frac{p_{A2}}{p_{A1}} \right)^{\frac{1}{\kappa}}
   $$



#### ⚖️ Bedeutung

* Wenn der Druck steigt ($p_{A2} > p_{A1}$), **verringert sich das spezifische Volumen** ($v_{A2} < v_{A1}$).
* Dadurch **nimmt die Masse** im betrachteten Volumen zu.
* Das Verhältnis hängt nur von der **Druckänderung** und dem **Isentropenexponenten κ** ab.



#### ✅ Zusammenfassung

Diese Gleichung beschreibt den Zusammenhang:

* **Höherer Druck → kleineres spezifisches Volumen → größere Masse im selben Raum.**
* Gültig nur für **isentropen (adiabaten + reversiblen)** Prozess eines idealen Gases.

#### 💡 Hinweis

Merke:
$$
p_1 v_1^{\kappa} = p_2 v_2^{\kappa} \quad \Rightarrow \quad \frac{v_1}{v_2} = \left( \frac{p_2}{p_1} \right)^{1/\kappa}
$$
Diese Relation taucht oft bei **Verdichtern oder Expansionsvorgängen** auf.

## 📘 Kreisprozesse

### 🧩 Formen von Kreisprozessen

* Ein **Kreisprozess** ist eine Abfolge von Zustandsänderungen, bei der das System nach einem Zyklus in den Ausgangszustand zurückkehrt.
* Wichtig: Am Ende gilt immer
  $$
  \Delta U = 0 \quad \Rightarrow \quad Q_{zu} + W = 0
  $$
  → Die zugeführte Wärme wird vollständig in Arbeit und Abwärme umgesetzt.

Typische Prozessbausteine:

* Isotherm: $T = \text{konst.}$
* Isobar: $p = \text{konst.}$
* Isochor: $V = \text{konst.}$
* Isentrop: $s = \text{konst.}$

Beispiele für Kreisprozesse:

* Carnot-Prozess (ideal, max. Wirkungsgrad)
* Otto-Prozess (Ottomotor)
* Diesel-Prozess
* Joule-Brayton-Prozess (Gasturbine)

### 🌡️ Beispiel – Carnot-Prozess

Ablauf:

1. **Isotherme Expansion** bei hoher Temperatur $T_H$: System nimmt Wärme $Q_H$ auf und verrichtet Arbeit.
2. **Isentrope Expansion**: Temperatur sinkt von $T_H$ auf $T_C$, System verrichtet Arbeit.
3. **Isotherme Kompression** bei niedriger Temperatur $T_C$: Wärme $Q_C$ wird an die Umgebung abgeführt.
4. **Isentrope Kompression**: Temperatur steigt von $T_C$ zurück auf $T_H$, Arbeit wird zugeführt.

Wirkungsgrad:
$$
\eta_C = 1 - \frac{T_C}{T_H}
$$

### ⚖️ Wichtige Größen

* **Wirkungsgrad allgemein:**
  $$
  \eta = \frac{|W|}{Q_{zu}} = 1 - \frac{Q_{ab}}{Q_{zu}}
  $$

* **Arbeit im Zyklus:**
  $$
  W = \oint p , dV
  $$
  → Fläche im $pV$-Diagramm.

* **Wärmeübertragung:** Prozessabhängig (isotherm, isobar, isochor, isentrop).

### 📏 Schritte zur Berechnung eines Kreisprozesses

1. **Zustandspunkte bestimmen** (Druck, Volumen, Temperatur, ggf. Entropie).
2. **Prozessgesetze anwenden** (z.B. $pV^\kappa = \text{konst.}$ für isentrop, $pV = RT$ für ideal).
3. **Arbeit und Wärme je Prozessabschnitt berechnen**:

   * isotherm: $W = nRT \ln \tfrac{V_2}{V_1}$
   * isobar: $W = p (V_2 - V_1)$
   * isochor: $W=0$, $Q=\Delta U$
   * isentrop: $W = \tfrac{p_2 V_2 - p_1 V_1}{1-\kappa}$
4. **Gesamtarbeit berechnen** (Summe der Teilprozesse).
5. **Gesamtwärmefluss berechnen**.
6. **Wirkungsgrad berechnen** mit $\eta = \tfrac{W}{Q_{zu}}$.

### ✅ Zusammenfassung

* Kreisprozesse wandeln Wärme in Arbeit (oder umgekehrt, bei Kältemaschinen).
* Am Ende ist die innere Energie unverändert, Bilanz hängt nur von $Q$ und $W$ ab.
* Wichtige Größen: $W = \oint p dV$, $\eta = 1 - \tfrac{Q_{ab}}{Q_{zu}}$.
* Carnot-Prozess liefert den theoretisch maximalen Wirkungsgrad $\eta_C = 1 - \tfrac{T_C}{T_H}$.
* Praktische Prozesse (Otto, Diesel, Joule-Brayton) nutzen Näherungen, die technisch realisierbar sind.

### 💡 Hinweis

Bei Aufgaben zu Kreisprozessen immer:

* **Prozessart erkennen** (isobar, isochor, isentrop, isotherm).
* **Gasgleichung anwenden** für Zustandsgrößen.
* **Wärme und Arbeit pro Schritt berechnen**, dann summieren.

## 📘 Kreisprozesse – Wirkungsgrade und Leistungszahlen

### 🧩 Thermischer Wirkungsgrad

* Allgemeine Definition:

$$
\eta = \frac{|W|}{Q_z} = 1 - \frac{Q_a}{Q_z}
$$

* $Q_z$: zugeführte Wärme
* $Q_a$: abgeführte Wärme
* $W$: im Zyklus gewonnene Arbeit



### 🌡️ Carnot-Wirkungsgrad

* Für den idealen, reversiblen Carnot-Prozess:

$$
\eta_C = 1 - \frac{T_a}{T_z}
$$

* $T_z$: Temperatur des Wärmereservoirs (hoch)
* $T_a$: Temperatur des Kältereservoirs (niedrig)

→ Kein realer Prozess kann effizienter sein als der Carnot-Prozess.


### ⚖️ Leistungszahlen

**Kältemaschine:**

$$
\varepsilon_K = \frac{Q_z}{W} = \frac{Q_z}{Q_a - Q_z}
$$

* Für den Carnot-Prozess:

$$
\varepsilon_K = \frac{T_z}{T_a - T_z}
$$

**Wärmepumpe:**

$$
\varepsilon_W = \frac{Q_a}{W} = \varepsilon_K + 1
$$

* Beispiel:
  Bei $T_z = 273K$, $T_a = 293K$:
  $\varepsilon_K = 13.7$, $\varepsilon_W = 14.7$.



### 📏 Verknüpfungen

* **Zusammenhänge:**

  * Wärmekraftmaschine → Wirkungsgrad $\eta$
  * Kältemaschine → Leistungszahl $\varepsilon_K$
  * Wärmepumpe → Leistungszahl $\varepsilon_W = \varepsilon_K + 1$

* **Grenze:**
  Alle realen Prozesse $\eta_{real} < \eta_C$.



### ✅ Zusammenfassung

* Der thermische Wirkungsgrad beschreibt, wie viel Arbeit aus der zugeführten Wärme entsteht.
* Der Carnot-Wirkungsgrad gibt die theoretische Obergrenze vor.
* Kältemaschinen und Wärmepumpen werden über Leistungszahlen bewertet, die größer als 1 sein können.
* Für gleiche Temperaturdifferenzen gilt: Wärmepumpen sind günstiger für Heizzwecke, Kältemaschinen für Kühlung.



### 💡 Hinweis

👉 Beim Rechnen immer auf **Vorzeichenkonvention** achten: $Q_z > 0$ (Zufuhr), $Q_a > 0$ (Abfuhr).
👉 Leistungszahlen sind stark temperaturabhängig – Herstellerangaben beziehen sich oft auf bestimmte Standardbedingungen.

## 📘 Entropie und 2. Hauptsatz

### 🧩 Zentrale Inhalte

* **Definition der Entropie**

  * Für einen reversiblen Prozess:
    $$
    dS = \frac{\delta Q_\text{rev}}{T}
    $$
  * Entropie ist eine Zustandsgröße (nicht prozessabhängig).

* **Clausius-Ungleichung**

  * Für einen beliebigen Kreisprozess:
    $$
    \oint \frac{\delta Q}{T} \leq 0
    $$
  * Gleichheit für reversible, Striktheit für irreversible Prozesse.

* **2. Hauptsatz**

  * Die Entropie eines abgeschlossenen Systems kann **nicht abnehmen**:
    $$
    \Delta S \geq 0
    $$
  * Nur bei reversiblen Prozessen bleibt sie konstant.

* **Quasistatische adiabate Zustandsänderung**

  * Für ideale Gase gilt:
    $$
    \delta Q = 0 \quad \Rightarrow \quad ds = 0
    $$
  * ⇒ **Adiabatisch + quasistatisch = isentrop**.

* **Reversibel vs. irreversibel**

  * Reversibel: idealisierte Prozesse ohne Reibung, Wärmeleitung, Dissipation.
  * Irreversibel: reale Prozesse mit Entropiezunahme (z.B. Wärmeleitung, Reibung, Mischung).

* **Entropiezunahme**

  * Irreversible Prozesse führen zur Entropieerhöhung.
  * Beispiel: Mischung zweier Gase bei gleichem Druck und Temperatur → **Mischungsentropie**.

* **Entropie und Unordnung**

  * Entropie = Maß für mikroskopische Unordnung.
  * Natürliche Prozesse streben Zuständen größerer Unordnung zu.

* **Entropiestrom und Entropieproduktion**

  * Entropiestrom durch Wärmeaustausch: $Q/T$.
  * Entropieproduktion tritt nur bei irreversiblen Prozessen auf und ist stets positiv.

### 🌡️ Wichtiges Beispiel

**Wärmeleitung zwischen zwei Reservoirs**

* Wärmefluss $Q$ von $T_1$ nach $T_2$ mit $T_1 > T_2$.
* Entropieänderung:
  $$
  \Delta S = -\frac{Q}{T_1} + \frac{Q}{T_2} > 0
  $$
* Zeigt klar: Wärmeleitung ist irreversibel.

### 📏 Wichtige Formeln und Schritte

1. **Definition:** $dS = \frac{\delta Q_\text{rev}}{T}$
2. **Clausius-Ungleichung:** $\oint \frac{\delta Q}{T} \leq 0$
3. **2. HS für abgeschlossenes System:** $\Delta S \geq 0$
4. **Isentrop:** $ds = 0 ;;\Rightarrow;; p v^\kappa = \text{const.}$
5. **Irreversibilität:** $\sigma > 0$ (Produktionsterm in Bilanz).

### ✅ Zusammenfassung

* Kapitel 4 verbindet die mathematische Definition von Entropie mit dem 2. Hauptsatz.
* Entropie ist eine **Zustandsgröße**, die nur bei reversiblen Prozessen konstant bleibt.
* In abgeschlossenen Systemen nimmt Entropie nie ab.
* Wärmeleitung, Reibung, Mischprozesse erzeugen Entropie.
* Entropie ist eng mit dem Konzept von **Unordnung** und **Irreversibilität** verbunden.

* Entropieänderung berechnen mit $\Delta S = \int \tfrac{\delta Q_\text{rev}}{T}$.
* Bei Kombination mehrerer Teilsysteme immer **Gesamtentropie** betrachten.

## 📘 Wärmetransport

### 🧩 Grundformen des Wärmetransports

1. **Wärmeleitung (Fouriersches Gesetz)**

   * Wärmestromdichte proportional zum Temperaturgradienten:
     $$
     \dot q = - \lambda , \frac{\partial T}{\partial x}
     $$
   * λ = Wärmeleitfähigkeit [W/mK]
   * Verallgemeinerung:
     $$
     \vec q = -\lambda , \nabla T
     $$
   * Wärmeleitgleichung (zeitabhängig):
     $$
     \frac{\partial T}{\partial t} = a , \nabla^2 T, \quad a = \frac{\lambda}{\rho c_p}
     $$

2. **Konvektion (Wärmeübergang)**

   * Mischung aus Wärmeleitung und Energietransport durch Strömung.
   * Unterscheidung:

     * **erzwungene Konvektion** (durch Pumpe, Gebläse)
     * **freie Konvektion** (durch Auftrieb im Schwerefeld)
   * Modellgleichung:
     $$
     \dot q = \alpha (T - T_m)
     $$
     mit Wärmeübergangskoeffizient $\alpha$ [W/m²K].
   * Bestimmung über **Nußelt-Zahl**:
     $$
     Nu = \frac{\alpha L}{\lambda}
     $$

### 🌡️ Anwendungen und Beispiele

* **Stationäre Wärmeleitung durch Wände/Zylinder**:

  * Temperaturverlauf bei Zylinderkoordinaten:
    $$
    T(r) = a \ln r + b
    $$

* **Beispiel Heißdampfleitung**:

  * Bestimmung des radialen Temperaturverlaufs in der Isolierung.
  * Berechnung des Wärmeverlusts pro Längeneinheit.

* **Beispiel Doppelverglasung/Wand**:

  * Berechnung thermischer Widerstände und Gesamtwärmestroms.
  * Verwendung von Wärmeleitfähigkeit und Wärmeübergangskoeffizienten.

### ⚖️ Wichtige Größen

* **Wärmeleitfähigkeit λ** (stoffabhängig, Temperaturabhängigkeit beachten).
* **Temperaturleitfähigkeit a = λ / (ρ c_p)** → Maß für Ausbreitungsgeschwindigkeit von Temperaturänderungen.
* **Wärmeübergangskoeffizient α** (abhängig von Strömung, Geometrie, Stoff).
* **Nußelt-Zahl Nu** zur dimensionslosen Beschreibung des Wärmeübergangs.

### 📏 Verknüpfungen

* Zusammenhang zu **Entropieproduktion**: Wärmeleitung und Konvektion sind **irreversible Prozesse**.
* Wärmeleitung → analog zu **Diffusion** von Teilchen.
* Wärmetransport bildet Basis für **Wärmedämmung, Heiztechnik, Kühlprozesse**.

### ✅ Zusammenfassung

Kapitel 5 behandelt die **Grundformen des Wärmetransports**:

* **Leitung**: Fourier-Gesetz, Differentialgleichung, Beispiele ebene Wand & Zylinder.
* **Konvektion**: Wärmeübergang an Oberflächen, Modell über Wärmeübergangskoeffizient & Nußelt-Zahl.
* **Anwendungen**: Wärmedämmung, Isolierung von Rohren, Fenster & Gebäudewände.

### 💡 Hinweis

Für Rechnungen im technischen Bereich wird oft das **Widerstandsmodell** genutzt:
$$
\dot Q = \frac{\Delta T}{R_{th}}
$$
mit $R_{th}$ = thermischer Widerstand, zusammengesetzt aus Leit- und Übergangswiderständen.

## 📘 Wärmeleitung und Wärmestrom – Übersicht

### 🧩 Grundbegriffe

| Symbol | Bedeutung                 | Vergleich / Analogie                           |
| ------ | ------------------------- | ---------------------------------------------- |
| **a**  | Temperaturleitfähigkeit   | Maß für Wärmeausbreitung im Material           |
| **α**  | Wärmeübergangskoeffizient | wie elektrischer Übergangswiderstand           |
| **q̇** | Wärmestromdichte (W/m²)   | analoge Größe zum elektrischen Stromdichtefeld |
| **Q̇** | Wärmestrom (W)            | analog zum elektrischen Strom I                |
| **A**  | Fläche (m²)               | Querschnitt, über den Wärme fließt             |
| **ΔT** | Temperaturunterschied     | analoge Größe zur Spannung ΔU                  |
| **λ**  | Wärmeleitfähigkeit (W/mK) | analog zur elektrischen Leitfähigkeit          |
| **R**  | Wärmewiderstand (K/W)     | analog zum elektrischen Widerstand R = U/I     |


### 🌡️ Wärmeübertragung und -leitung

**Für Konvektion:**

$$
\dot{q} = \alpha , \Delta T
$$

**Für Wärmeleitung:**

$$
\dot{q} = - \lambda , \frac{\partial T}{\partial x}
$$

*Das Minuszeichen zeigt: Wärme fließt von warm nach kalt.*

Im stationären Fall gilt:

$$
\dot{Q} = \text{konstant}
$$



### ⚖️ Wärmewiderstände

**Allgemein:**

$$
R = \frac{1}{\alpha} \quad \text{(für Konvektion)}
$$

$$
R = \frac{L}{\lambda} \quad \text{(für Wärmeleitung)}
$$

Kombination von Widerständen (wie in der Elektrizitätslehre):

* Serienschaltung: ( $R_{ges} = R_1 + R_2 + \dots $ )
* Parallelschaltung: ( $\tfrac{1}{R_{ges}} = \tfrac{1}{R_1} + \tfrac{1}{R_2} + \dots $ )



### 📏 Berechnung des Wärmestroms

Gesamtwärmestrom:

$$
\dot{Q} = \dot{q} \cdot A
$$

Mit Fourier’s Gesetz:

$$
\dot{q} = - \lambda \frac{\partial T}{\partial x}
$$

und damit

$$
\dot{Q} = - \lambda A \frac{\partial T}{\partial x}
$$

Für den stationären Fall (konstante Temperaturdifferenz über Länge L):

$$
\dot{Q} = \frac{\lambda A}{L} \cdot \Delta T = \frac{\Delta T}{R_{th}}
$$



### ✅ Zusammenfassung

* Wärmeübertragung funktioniert analog zum elektrischen Stromfluss.
* Temperaturunterschied ⇔ Spannung, Wärmestrom ⇔ Strom.
* Stationär gilt immer ein konstanter Wärmestrom.
* Der Wärmewiderstand ergibt sich je nach Mechanismus:

  * **Leitung:** ( $R = \frac{L}{\lambda A}$ )
  * **Konvektion:** ( $R = \frac{1}{\alpha A}$ )
* Kombination erfolgt wie bei elektrischen Widerständen.



### 💡 Hinweis

Die roten Markierungen im Bild („**Const. im stationären Fall**“) betonen,
dass sich der Wärmestrom über alle Schichten **nicht ändert**, solange keine
Wärmequellen oder -senken innerhalb des Systems vorhanden sind.

## 2. **Warum ist $\dot Q$ immer gleich?**

Weil wir von einem **stationären** Fall ausgehen:

* Es wird keine Wärme zwischengespeichert.
* Was **innen** in die erste Schicht hineingeht, muss exakt durch jede Schicht hindurch weiter nach außen fließen.
* Das heißt: durch jede Schicht gilt derselbe Wärmestrom (\dot Q).

# 👉 HS der TD

## 📘 Übersichtstabelle

### 🧩 Übersicht

| Hauptsatz      | Formulierung                                                                                             | Mathematische Darstellung                                                     | Verwendung                                                                        |
| -------------- | -------------------------------------------------------------------------------------------------------- | ----------------------------------------------------------------------------- | --------------------------------------------------------------------------------- |
| **Nullter HS** | Systeme im thermischen Gleichgewicht mit einem dritten stehen auch miteinander im Gleichgewicht          | –                                                                             | Definition und Messung der Temperatur                                             |
| **Erster HS**  | Energieerhaltung: Änderung der inneren Energie = Wärme + Arbeit                                          | $$U_2 - U_1 = Q_{12} + W_{12}$$ $$dU = \delta Q + \delta W$$ $$du = dq + dw$$ | Energiebilanzen in geschlossenen und offenen Systemen                             |
| **Zweiter HS** | Wärme kann nicht vollständig in Arbeit umgewandelt werden; Entropie nimmt bei irreversiblen Prozessen zu | $$dS = \frac{\delta Q_\text{rev}}{T}$$ $$\Delta S \geq 0$$                    | Richtung realer Prozesse, Grenzen von Wirkungsgraden, Kältemaschinen, Wärmepumpen |
| **Dritter HS** | Der absolute Nullpunkt $T=0$ K kann nicht durch endliche Prozesse erreicht werden                        | $$\lim_{T \to 0} S = \text{konst.}$$                                          | Definition der absoluten Temperaturskala, Verhalten bei tiefen Temperaturen       |


### 🌡️ Wichtige Formeln

* Nullter HS: keine eigene Gleichung, Grundlage für
  $$T_A = T_B \quad \text{(wenn A und B im thermischen Gleichgewicht)}$$

* Erster HS (geschlossene Systeme):
  $$
  \Delta U = Q + W
  $$

  Spezifisch:
  $$
  du = dq + dw
  $$

  Differentiell:
  $$
  dU = \delta Q + \delta W
  $$

* Zweiter HS (Entropie):
  $$
  dS = \frac{\delta Q_\text{rev}}{T}, \quad \Delta S \geq 0
  $$

* Dritter HS (Unerreichbarkeit des Nullpunkts):
  $$
  \lim_{T \to 0} S = \text{konst.}
  $$


### ⚖️ Anwendung

* Nullter HS → Temperaturmessung, Thermometer, Definition thermisches Gleichgewicht.
* Erster HS → Energiebilanzen von Motoren, Turbinen, Wärmeaustauschern.
* Zweiter HS → Bestimmung maximal möglicher Wirkungsgrade (Carnot), Erklärung irreversibler Prozesse.
* Dritter HS → Verhalten von Stoffen nahe 0 K, kryotechnische Anwendungen.

### ✅ Zusammenfassung

Die Hauptsätze bauen logisch aufeinander auf:

1. **Nullter HS** definiert Temperatur.
2. **Erster HS** beschreibt Energieerhaltung.
3. **Zweiter HS** legt Prozessrichtungen und Effizienzgrenzen fest.
4. **Dritter HS** bestimmt die untere Grenze der Temperaturskala.

### 💡 Hinweis

Bei Berechnungen immer unterscheiden:

* Geschlossene Systeme → $\Delta U = Q + W$
* Offene Systeme → zusätzlich Massen- und Enthalpieströme berücksichtigen.

### 2.

* Zeitlich abgeleitete Form des 1. Hauptsatzes (geschlossenes, ruhendes System, quasistatisch):

  $$
  \dot U = \dot Q - p,\dot V
  \qquad\Rightarrow\qquad
  m,c_v,\dot T = \dot Q - p,\dot V
  $$
  für ein ideales Gas mit $u=m c_v T$. 

* Für den Kolben mit Feder (System = Gas, Wärmezufuhr $\dot Q = P$):

  $$
  m,c_v,\dot T = P - p,\dot V

* Alternative Schreibweise (Aufspaltung der Volumenarbeit in Umgebungs- und Federanteil):

  $$
  m,c_v,\dot T = P - p_0,\dot V - k,(x-x_0),\dot x
  $$
  denn $p,\dot V = p_0,\dot V + k(x-x_0)\dot x$.

### 🌡️ Herleitung

* Ausgangspunkt (differenzielle Form des 1. HS für einfache Systeme, quasistatisch):

  $$
  dU = dQ - p,dV.
  $$
  Ableiten nach der Zeit liefert
  $$
  \dot U = \dot Q - p,\dot V.
  $$


* Für ideales Gas gilt $u=u(T)$, somit $U = m,u = m,c_v,T$ und daher
  $$
  \dot U = m,c_v,\dot T.
  $$


* In der Aufgabe ist $\dot Q = P$ gegeben, also
  $$
  m,c_v,\dot T = P - p,\dot V.
  $$

* Zusammenhang zur Kolbenbewegung: $V=A x \Rightarrow \dot V = A \dot x$. Bei quasistatischer Bewegung ist die Kolbenkraft im Gleichgewicht mit Umgebungsdruck und Federkraft, daher
  $$
  p = p_0 + \frac{k}{A},(x-x_0).
  $$

* Einsetzen:
  $$
  m,c_v,\dot T

  # P - \Bigl(p_0 + \frac{k}{A}(x-x_0)\Bigr),A,\dot x

  P - p_0 A \dot x - k(x-x_0)\dot x.
  $$

### ⚖️ Gaskonstanten

* Ideales Gas im Zylinder:
  $$
  p,V = m,R,T,\qquad V=A,x.
  $$
  Damit kannst du $p$ oder $T$ bei Bedarf eliminieren. 

* Alternative enthalpie-basierte Form (nur zur Einordnung):
  $$
  dH = dQ + V,dp
  \quad\Rightarrow\quad
  \dot H = \dot Q + V,\dot p,
  $$
  für quasistatische Prozesse; hier aber nicht nötig, da Volumenarbeit direkt mit $p,\dot V$ erfasst wird. 

### 📏 Verknüpfungen

* Kopplung Thermo–Mechanik (praktisch für spätere Eliminierung von $T$):
  $$
  m,c_v,\dot T = P - p,A,\dot x,\qquad
  p = \frac{m R T}{A x} = p_0 + \frac{k}{A}(x-x_0).
  $$
  Erste Gleichung = 1. HS in Zeitableitung; zweite/ dritte Gleichung = Gasgesetz bzw. Kraftgleichgewicht.

  $$
  k(x-x_0)\dot x = \frac{d}{dt}!\left(\tfrac12 k(x-x_0)^2\right),

  sodass
  $$
  m,c_v,\dot T = P - p_0 A \dot x - \frac{d}{dt}!\left(\tfrac12 k(x-x_0)^2\right).
  $$
  Diese Form zeigt, dass die Wärmezufuhr in innere Energie, Umgebungsausdehnungsarbeit und Federenergie geht.

* Gesuchter 1. Hauptsatz in zeitlich abgeleiteter Form (für das Gas):

  m,c_v,\dot T = P - p,\dot V,

  $$
  m,c_v,\dot T = P - \Bigl(p_0 + \frac{k}{A}(x-x_0)\Bigr),A,\dot x.
