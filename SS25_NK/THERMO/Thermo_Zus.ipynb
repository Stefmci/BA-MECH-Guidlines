## üìò Ideales Gasgesetz

### üß© Formen des Gasgesetzes

| Form                        | Formel              | Gr√∂√üen                |
| --------------------------- | ------------------- | --------------------- |
| **Spezifisch**              | $p , v = R , T$     | $v = \tfrac{1}{\rho}$ |
| **Massenbezogen**           | $p , V = m , R , T$ | $m =$ Masse           |
| **Molbezogen (universell)** | $p , V = n , R , T$ | $n =$ Stoffmenge      |


### üå°Ô∏è Herleitung

* Boyle‚ÄìMariotte:
  $$
  p , v \big|_{T} = \text{konst.}
  $$
* Gay-Lussac:
  $$
  v \big|_{p} \sim T
  $$
* Kombination:
  $$
  p , v = R , T
  $$


### ‚öñÔ∏è Gaskonstanten

* Universelle Gaskonstante:
  $$
  R = 8{,}314 \ \tfrac{\text{J}}{\text{mol} \cdot \text{K}}
  $$
* Spezifische Gaskonstante:
  $$
  R_s = \tfrac{R}{M}
  $$

  Beispiel Luft:
  $$
  R_s = 286{,}7 \ \tfrac{\text{J}}{\text{kg} \cdot \text{K}}, \quad M \approx 28{,}96 \ \tfrac{\text{g}}{\text{mol}}
  $$



### üìè Verkn√ºpfungen

* Formen: spezifisch, massenbezogen, molbezogen
* Konstanten:
  $$
  R = 8{,}314 \ \tfrac{\text{J}}{\text{mol} \cdot \text{K}}, \quad R_s = \tfrac{R}{M}
  $$
* Beispiel Luft:
  $$
  R_s = 286{,}7 \ \tfrac{\text{J}}{\text{kg} \cdot \text{K}}
  $$
* Wichtige Beziehungen:
  $$
  p = \rho \cdot R \cdot T,  \quad c_p - c_v = R, \quad \kappa = \tfrac{c_p}{c_v}
  $$



### üí° Hinweis

* Drei gleichwertige Schreibweisen:
  $$
  p \cdot v = R \cdot T, \quad p \cdot V = m \cdot R \cdot T, \quad p \cdot V = n \cdot R \cdot T
  $$
* Universell = unabh√§ngig vom Stoff
* Spezifisch = stoffabh√§ngig (√ºber Molmasse)
* N√§herung gilt gut f√ºr ideale Gase bei normalen Dr√ºcken und Temperaturen



### ‚úÖ Zusammenfassung

* **Formen:** spezifisch, massenbezogen, molbezogen
* **Konstanten:** $R = 8{,}314 \ \tfrac{\text{J}}{\text{mol} \cdot \text{K}}, \quad R_s = \tfrac{R}{M}$
* **Beispiel Luft:** $R_s = 286{,}7 \ \tfrac{\text{J}}{\text{kg} \cdot \text{K}}$
* **Wichtige Verkn√ºpfungen:** $p = \rho R T, \quad c_p - c_v = R, \quad \kappa = \tfrac{c_p}{c_v}$

## üìò Thermodynamische Prozesse 

### üß© Hauptprozesse

| Prozess      | Bedingung      | Hauptformel                                | Bemerkung               |
| ------------ | -------------- | ------------------------------------------ | ----------------------- |
| **Isotherm** | ( T = konst. ) | ( p V = \text{konst.} )                    | W√§rme = Arbeit          |
|              |                | ( W = n R T \ln \tfrac{V_2}{V_1} )         |                         |
| **Isochor**  | ( V = konst. ) | ( W = 0 )                                  | W√§rme = ŒîU              |
|              |                | ( Q = \Delta U )                           |                         |
| **Isobar**   | ( p = konst. ) | ( W = p (V_2 - V_1) )                      | W√§rme = ŒîH              |
|              |                | ( Q = \Delta H )                           |                         |
| **Isentrop** | ( S = konst. ) | ( p V^\kappa = \text{konst.} )             | reversibel + adiabat    |
|              |                | ( T V^{\kappa-1} = \text{konst.} )         |                         |
|              |                | ( W = \tfrac{p_2 V_2 - p_1 V_1}{1-\kappa}) |                         |
| **Adiabat**  | ( Q = 0 )      | ( \Delta U = W )                           | W√§rmestrom = 0          |


### üå°Ô∏è Allgemeine Formeln

* Arbeit allgemein: ( W = \int p , dV )
* W√§rme allgemein: ( Q = \int T , dS )
* Erster Hauptsatz: ( \Delta U = Q + W )


### üí° Hinweis

* Isotherm: Temperatur konstant, ( Q = W )
* Isochor: Volumen konstant, ( Q = \Delta U )
* Isobar: Druck konstant, ( Q = \Delta H )
* Isentrop: Entropie konstant, ( p V^\kappa = konst. )
* Adiabat: kein W√§rmeaustausch, ( \Delta U = W )


### ‚úÖ Zusammenfassung

* Prozesse = Zustands√§nderungen, eingeteilt nach Konstanz einer Gr√∂√üe
* Allgemeine Form: ( p V^n = konst. )
* Arbeit und W√§rme √ºber Integrale ( W = ‚à´p dV, Q = ‚à´T dS )
* Isentrop = Spezialfall von adiabatem, wenn reversibel

## üìò p/v und T-S Diagramm

```python
# Volumen-Bereich
V = np.linspace(0.5, 3.0, 100)
n = 1.0   # mol
R = 8.314
T_iso = 1
p_iso = n*R*T_iso/V

# Isotherm (T = konst.)
p_isotherm = p_iso

# Isobar (p = konst.)
p_isobar = np.ones_like(V)*1.0

# Isochor (V = konst.)
V_isochor = np.ones(200)*1.0
p_isochor = np.linspace(0.5,15,200)

# Isentrop (p*V^kappa=konst.)
kappa = 1.4
p_isentrop = p_iso[0]*V[0]**kappa / V**kappa

# Plot p-V Diagramm
plt.figure(figsize=(8,6))
plt.plot(V, p_isotherm, label="Isotherm (T=konst.)")
plt.plot(V, p_isentrop, label="Isentrop (S=konst.)")
plt.plot(V, p_isobar, label="Isobar (p=konst.)")
plt.plot(V_isochor, p_isochor, label="Isochor (V=konst.)")

plt.xlabel("Volumen V [m¬≥]")
plt.ylabel("Druck p [bar, normiert]")
plt.title("p-V-Diagramm: Thermodynamische Prozesse")
plt.legend()
plt.grid(True)
plt.show()
```

```python
s = np.linspace(0, 3, 200)
T0 = 300

# Isotherm: T = konst
T_isotherm = np.ones_like(s)*T0

# Isobar: T steigt st√§rker mit Entropie, leichte Kr√ºmmung
T_isobar = T0 + 120*(np.sqrt(s+1)-1)

# Isochor: T steigt noch st√§rker, deutlichere Kr√ºmmung
T_isochor = T0 + 300*(np.sqrt(s+1)-1)

# Isentrop: s = konst
s_isentrop = np.ones(200)*1.5
T_isentrop = np.linspace(250, 600, 200)

# Plot T-s Diagramm
plt.figure(figsize=(8,6))
plt.plot(s, T_isotherm, label="Isotherm (T=konst.)", linewidth=2)
plt.plot(s, T_isobar, label="Isobar (p=konst.)", linewidth=2)
plt.plot(s, T_isochor, label="Isochor (V=konst.)", linewidth=2)
plt.plot(s_isentrop, T_isentrop, label="Isentrop (S=konst.)", linewidth=2)

plt.xlabel("Entropie s [kJ/(kg K)] (schematisch)")
plt.ylabel("Temperatur T [K]")
plt.title("T-s-Diagramm: Thermodynamische Prozesse")
plt.legend()
plt.grid(True)
plt.show()
```

## üìò Erster Hauptsatz der Thermodynamik

### üß© Allgemeine Form

Der **1. Hauptsatz** beschreibt die **Energieerhaltung**:

$$
\Delta U = Q + W
$$

mit

* $U$ = innere Energie
* $Q$ = zugef√ºhrte W√§rme
* $W$ = zugef√ºhrte Arbeit (inkl. Volumen√§nderungsarbeit, technische Arbeit etc.)


### üå°Ô∏è Geschlossene Systeme

F√ºr ein ruhendes, geschlossenes System:

$$
dU = \delta Q + \delta W
$$

* **Differentiell, spezifisch:**
  $$
  du = \delta q + \delta w
  $$

* **Bei quasistatischer Zustands√§nderung:**
  $$
  dU = \delta Q - p \cdot dV
  $$

* **Sonderf√§lle:**

  * Isochor ($dV = 0$):
    $$dU = \delta Q$$
  * Isobar:
    $$dH = \delta Q$$


### ‚öñÔ∏è Offene Systeme

Hier muss zus√§tzlich der **Energieaustausch durch Massenstr√∂me** ber√ºcksichtigt werden.

Allgemein:

$$
dE = \delta Q + \delta W + d^{(m)}E
$$

mit $d^{(m)}E$ = Energie durch Massenstrom (Transport von innerer, kinetischer und potentieller Energie).

* Aufgeteilt:

$$
dU = \delta Q + \delta W_t + p_z v_z , dm_z - p_a v_a , dm_a + d^{(m)}U
$$

$z =$ Zustrom, $a =$ Abstrom.

* **Station√§rer Fall (Energiegleichung):**

$$
0 = \dot Q + \dot W_t + \sum \dot m \left(h + \frac{c^2}{2} + gz \right)*{\text{ein}} - \sum \dot m \left(h + \frac{c^2}{2} + gz \right)*{\text{aus}}
$$

Dies ist die **Energiegleichung f√ºr station√§re offene Systeme** (z. B. Turbine, Verdichter, W√§rmetauscher).


### üìè Abwandlungen

* **Differentialform:**
  $$
  dU = \delta Q + \delta W
  $$

* **Spezifisch (pro Masse):**
  $$
  du = \delta q + \delta w
  $$

* **Enthalpie-Form (f√ºr offene Systeme, station√§r):**
  $$
  \dot Q + \dot W_t = \sum \dot m , h_{\text{ein}} - \sum \dot m , h_{\text{aus}}
  $$



### ‚úÖ Zusammenfassung

* Geschlossen: **Innere Energie√§nderung = W√§rme + Arbeit**.
* Offenes System: **Zus√§tzlich Energie durch Massenstr√∂me (Enthalpie, kinetisch, potenziell)**.
* Station√§r: **Energieerhaltung in Bilanzgleichung (oft mit Enthalpie ausgedr√ºckt)**.

### üí° Hinweis

Im geschlossenen System √§ndern nur **$Q$ und $W$** die Energie.
Im offenen System kommt zus√§tzlich der **konvektive Transport** mit der Masse dazu ‚Üí deshalb wird oft mit **Enthalpie** gerechnet.

### üß© √úbersicht

## üìò Kreisprozesse ‚Äì Wirkungsgrade und Leistungszahlen

### üß© Thermischer Wirkungsgrad

* Allgemeine Definition:

$$
\eta = \frac{|W|}{Q_z} = 1 - \frac{Q_a}{Q_z}
$$

* $Q_z$: zugef√ºhrte W√§rme
* $Q_a$: abgef√ºhrte W√§rme
* $W$: im Zyklus gewonnene Arbeit



### üå°Ô∏è Carnot-Wirkungsgrad

* F√ºr den idealen, reversiblen Carnot-Prozess:

$$
\eta_C = 1 - \frac{T_a}{T_z}

* $T_z$: Temperatur des W√§rmereservoirs (hoch)
* $T_a$: Temperatur des K√§ltereservoirs (niedrig)

‚Üí Kein realer Prozess kann effizienter sein als der Carnot-Prozess.


### ‚öñÔ∏è Leistungszahlen

**K√§ltemaschine:**

$$
\varepsilon_K = \frac{Q_z}{W} = \frac{Q_z}{Q_a - Q_z}

* F√ºr den Carnot-Prozess:

* Beispiel:
  Bei $T_z = 273K$, $T_a = 293K$:
  $\varepsilon_K = 13.7$, $\varepsilon_W = 14.7$.



### üìè Verkn√ºpfungen

* **Zusammenh√§nge:**

  * W√§rmekraftmaschine ‚Üí Wirkungsgrad $\eta$
  * K√§ltemaschine ‚Üí Leistungszahl $\varepsilon_K$
  * W√§rmepumpe ‚Üí Leistungszahl $\varepsilon_W = \varepsilon_K + 1$

* **Grenze:**

### ‚úÖ Zusammenfassung

## üìò Entropie und 2. Hauptsatz

### üß© Zentrale Inhalte

* **Definition der Entropie**

  * F√ºr einen reversiblen Prozess:
    $$
    dS = \frac{\delta Q_\text{rev}}{T}
    $$
  * Entropie ist eine Zustandsgr√∂√üe (nicht prozessabh√§ngig).

* **Clausius-Ungleichung**

    $$

  * Gleichheit f√ºr reversible, Striktheit f√ºr irreversible Prozesse.

* **2. Hauptsatz**

    $$

* **Quasistatische adiabate Zustands√§nderung**

  * F√ºr ideale Gase gilt:
    $$
    \delta Q = 0 \quad \Rightarrow \quad ds = 0
    $$
  * ‚áí **Adiabatisch + quasistatisch = isentrop**.

* **Reversibel vs. irreversibel**

  * Reversibel: idealisierte Prozesse ohne Reibung, W√§rmeleitung, Dissipation.
  * Irreversibel: reale Prozesse mit Entropiezunahme (z.B. W√§rmeleitung, Reibung, Mischung).

* **Entropiezunahme**

  * Irreversible Prozesse f√ºhren zur Entropieerh√∂hung.
  * Beispiel: Mischung zweier Gase bei gleichem Druck und Temperatur ‚Üí **Mischungsentropie**.

* **Entropie und Unordnung**

  * Entropie = Ma√ü f√ºr mikroskopische Unordnung.
  * Nat√ºrliche Prozesse streben Zust√§nden gr√∂√üerer Unordnung zu.

* **Entropiestrom und Entropieproduktion**

  * Entropiestrom durch W√§rmeaustausch: $Q/T$.
  * Entropieproduktion tritt nur bei irreversiblen Prozessen auf und ist stets positiv.
