# Herleitung der Impulserhaltungsgleichung und Massenerhaltungsgleichung

Nicht ganz ‚Äì $\rho\,v^2\,A$ ist **kein Impuls** (also kein $\mathrm{kg\,m/s}$), sondern der **Impulsstrom** (bzw. die Momentenflussdichte) oder anders gesagt die **Kraft**, die dir das Fluid an diesem Querschnitt liefert.

## Herleitung

1. **Massenstrom**

   $$
     \dot m = \rho\,v\,A
     \quad\Bigl[\tfrac{\text{kg}}{\text{s}}\Bigr]
   $$

2. **Momentum¬≠strom**

   $$
     \dot m\;v \;=\; (\rho\,v\,A)\;v 
     = \rho\,v^2\,A
     \quad\Bigl[\tfrac{\text{kg}}{\text{s}}\times\!\tfrac{\text{m}}{\text{s}}
       = \tfrac{\text{kg\,m}}{\text{s}^2} = \text{N}\Bigr]
   $$

   Das ist genau ein **Kraft¬≠term** (Newton), den du in der Impuls¬≠bilanz als
   $\int\rho\,v_n\,v\,dA$ bekommst.

3. **Impuls vs. Impulsstrom**

   * **Impuls** $p$ eines Massen¬≠pakets ist $m\,v$ (Einheit kg¬∑m/s).
   * **Impulsstrom** bzw. **Kraft** ist die **√Ñnderung des Impulses pro Zeit**, also $\frac{dp}{dt} = F$.

Wenn du also $\rho\,v^2\,A$ mit einer Zeitspanne $\Delta t$ multiplizierst, bekommst du den **Impuls**

$$
I = F\,\Delta t = (\rho\,v^2\,A)\,\Delta t,
$$

der in dieser Zeit ins Volumen hinein- (oder hinaus-)transportiert wurde.



#  **Bernoulli-Gleichung**

## **Ideale Form (f√ºr inkompressible, reibungsfreie, station√§re Str√∂mung):**

$$

$$

### **Bedeutung der Terme:**

* **$p$**: **Statischer Druck** \[Pa]
  ‚Üí Druck, der senkrecht auf die Oberfl√§che wirkt (f√ºhlbar z.‚ÄØB. im Pitot-Rohr).
* **$\frac{1}{2} \rho v^2$**: **Dynamischer Druck** \[Pa]
  ‚Üí ‚ÄûBewegungsdruck‚Äú; Energieanteil durch Geschwindigkeit.
* **$\rho g z$**: **Hydrostatischer Druck** \[Pa]
  ‚Üí Potenzielle Energie pro Volumen durch Lage im Schwerefeld.

> **Interpretation:**
> Die Summe aus Druck-, Bewegungs- und Lageenergie bleibt entlang einer Stromlinie konstant, **wenn keine Verluste** auftreten.



##  **Erweiterung mit H√∂henunterschieden:**

$$
\frac{p_1}{\rho} + \frac{v_1^2}{2} + g z_1 = \frac{p_2}{\rho} + \frac{v_2^2}{2} + g z_2
$$



##  **Erweiterung: Bernoulli mit Verlusten (z.‚ÄØB. durch Reibung):**

### **Summenform (punktweise Betrachtung, z.‚ÄØB. Rohrnetz):**

$$
\frac{p_1}{\rho} + \frac{v_1^2}{2} + g z_1 = \frac{p_2}{\rho} + \frac{v_2^2}{2} + g z_2 + \Delta h_{\text{verluste}}

‚Üí **$\Delta h_{\text{verluste}}$** ist der **spezifische Energieverlust** \[m] z.‚ÄØB. durch Reibung, Formverluste, Turbulenz usw.

> ‚ö† Gilt auch in Form:

$$
p_1 + \frac{1}{2} \rho v_1^2 + \rho g z_1 = p_2 + \frac{1}{2} \rho v_2^2 + \rho g z_2 + \sum_i \zeta_i \cdot \frac{1}{2} \rho v_i^2

mit **$\zeta_i$** als **Verlustbeiwert** pro Element (z.‚ÄØB. Kr√ºmmer, Ventil, Verengung).



### **Integrale Form (entlang einer Stromlinie, kontinuierliche Verluste):**

$$
\int_{1}^{2} \left( \frac{dp}{\rho} + v \, dv + g \, dz \right) = -\int_{1}^{2} \frac{dE_{\text{verlust}}}{\rho}
$$

‚Üí Energieverluste werden hier als kontinuierlich verteilte Gr√∂√üen integriert.



## **Annahmen (idealer Fall):**

##  **Massenerhaltung (Kontinuit√§tsgleichung)**

###  **Differenzielle Form:**
